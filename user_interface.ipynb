```python
import tkinter as tk

def create_grid():
    global grid
    global buttons
    try:
        rows = int(entry_rows.get())
        cols = int(entry_cols.get())
        if rows > 0 and cols > 0:
            # Limpiar la ventana anterior si existe
            for row in buttons:
                for button in row:
                    button.destroy()
            grid = [[tk.IntVar() for _ in range(cols)] for _ in range(rows)]
            buttons = []
            for i in range(rows):
                row_buttons = []
                for j in range(cols):
                    button = tk.Button(ventana, width=2, height=1, bg="white", text=f"{column_labels[j]}{i+1}", command=lambda row=i, col=j: toggle_cell(row, col))
                    button.grid(row=i+1, column=j+1)
                    row_buttons.append(button)
                buttons.append(row_buttons)
        else:
            raise ValueError("Las dimensiones deben ser mayores que cero.")
    except ValueError as e:
        print(f"Error: {e}")

# Crear una instancia de la ventana principal
ventana = tk.Tk()
ventana.title("Gradilla Dinámica")

# Etiquetas para las letras de las columnas
column_labels = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Etiqueta para filas
tk.Label(ventana, text="Filas:").grid(row=0, column=0)
entry_rows = tk.Entry(ventana)
entry_rows.grid(row=0, column=1)

# Etiqueta para columnas
tk.Label(ventana, text="Columnas:").grid(row=1, column=0)
entry_cols = tk.Entry(ventana)
entry_cols.grid(row=1, column=1)

# Botón para crear la gradilla
create_button = tk.Button(ventana, text="Crear Gradilla", command=create_grid)
create_button.grid(row=2, columnspan=2)

# Listas para almacenar los valores de las casillas y los botones de la cuadrícula
grid = []
buttons = []

# Ejecutar el bucle principal de la aplicación
ventana.mainloop()
```
